```python
#Neuronal decoding of temperature signals in Caenorhabditis elegans
#Contributed by- Abhilasha Batra
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def sigmoid(t, a, tm, beta, b):
    y = a / (1 + np.exp(-beta*(t - tm))) + b
    return y

# Load data
data = np.genfromtxt('Kobayashi_Signal.csv', delimiter=',')
x_data, y_data = data[:, 0], data[:, 1]

# Initial guess for parameters
p0 = [max(y_data), np.median(x_data), 1, min(y_data)]

# Fit the data to the sigmoid model
popt, pcov = curve_fit(sigmoid, x_data, y_data, p0, method='dogbox')

# Print optimized parameters
print(f"Optimized parameters: {popt}")

# Generate values for the fitted curve
x_fit = np.linspace(min(x_data), max(x_data), 500)
y_fit = sigmoid(x_fit, *popt)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, 'o', color='darkblue', label='Experimental Data', markersize=6)
plt.plot(x_fit, y_fit, color='red', label='Sigmoid Fit', linewidth=3)
plt.xlabel('Time (sec)', fontsize=14)
plt.ylabel('Temperature ($^\circ$C)', fontsize=14)
plt.title('Sigmoidal Fit to Temperature Signal in Kobayashi et al.', fontsize=16)
plt.legend(loc='best', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
